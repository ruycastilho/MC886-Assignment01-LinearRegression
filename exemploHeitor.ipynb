```python
from sklearn.linear_model import SGDRegressor
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5,5,num=100)[:,None]
y = -0.5 + 2.2*x +0.3*x**3+ 2*np.random.randn(100,1)

plt.plot(x,y)
plt.show()
```

```python
print x.shape
model = SGDRegressor(n_iter=1000, penalty=None, learning_rate='constant', eta0=0.00001)
model.fit(x,y)
y_pred = model.predict(x)

plt.scatter(x,y)
plt.plot(x,y_pred,'r')
plt.show()
```

```python
x_new = np.hstack([x,x**2,x**3,x**4])

model = SGDRegressor(n_iter=100, learning_rate='constant', eta0=0.000001)
model.fit(x_new,y)

y_pred = model.predict(x_new)

plt.scatter(x,y)
plt.plot(x_new[:,0],y_pred,'r')
plt.show()
```
